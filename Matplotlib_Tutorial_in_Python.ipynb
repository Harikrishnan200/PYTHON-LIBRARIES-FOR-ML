```python
# import matplotlib library
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)  # Generate an array of 100 values from 0 to 10
y = np.sin(x)   # Generate corresponding sin values of x
z = np.cos(x)   # Generate corresponding cos values of x
```

```python
# Cosine wave
plt.plot(x, z)   # z is the cosine value of x
plt.show()
```

```python
# Adding title, x-axis and y-axis
plt.plot(x, y)
plt.title('Sine Wave')  # To set the title for the graph
plt.xlabel('Angle')     # To set a label to x axis as Angle
plt.ylabel('Sine value') # Tos set a lable to y axis as Sine value
plt.show()
```

```python

a = 2  # set the value of a

# Generate x values (only positive x values, since y^2 = 4ax is defined for x >= 0)
x = np.linspace(0, 10, 400)

# Compute the corresponding y values for the parabola y^2 = 4ax
y_positive = np.sqrt(4 * a * x)
y_negative = -np.sqrt(4 * a * x)

# Plot the graph with red dots
plt.plot(x, y_positive, 'r.', label='$y = \sqrt{4ax}$')
plt.plot(x, y_negative, 'r.', label='$y = -\sqrt{4ax}$')

plt.title('Parabola: $y^2 = 4ax$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Parameter 'a'
a = 1  # You can change this value to see different parabolas

# Generate x values (only positive x values, since y^2 = 4ax is defined for x >= 0)
x = np.linspace(0, 10, 400)

# Compute the corresponding y values for the parabola y^2 = 4ax
y_positive = np.sqrt(4 * a * x)
y_negative = -np.sqrt(4 * a * x)

# Plot the graph with red dots
plt.plot(x, y_positive, 'r.', label='$y = \sqrt{4ax}$')  # \sqrt{4ax} is the latex syntax of sqrt.(4ax)
plt.plot(x, y_negative, 'r.', label='$y = -\sqrt{4ax}$')

plt.title('Parabola: $y^2 = 4ax$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() # The plt.legend() function displays these labels in a legend, making it clear which curve corresponds to which mathematical expression.
plt.grid(True)
plt.show()


# Mathematical Symbols: The backslash \ is used to denote a LaTeX command. For example, \sqrt is the LaTeX command for the square root symbol.
# LaTeX Syntax: The entire expression is enclosed in dollar signs $...$ to indicate that it should be interpreted as a LaTeX math expression.
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-' , label='linear')
plt.plot(x, np.cos(x), 'r--', label='quadratic')
plt.plot(x, np.tan(x), 'b.', label='cubic')
#plt.plot(x, np.tan(x), label='tan')
plt.legend()
plt.show()
```

```python
x = np.linspace(-5, 5, 50)

# Plot different functions with labels
plt.plot(x, x, label='linear')        # Linear function
plt.plot(x, x**2, label='quadratic')  # Quadratic function
plt.plot(x, x**3, label='cubic')      # Cubic function

# Adding a title and labels to the axes
plt.title('Different Functions')
plt.xlabel('x')
plt.ylabel('y')

# Display the legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)   # To generate the bar
plt.xlabel('Programming Languages')
plt.ylabel('No. of Students')
plt.title('Students in Programming Languages')
plt.show()
```

```python
# Just for an example
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.7, 0.7])  # Axes takes up 70% of the figure size
ax2 = fig.add_axes([0.2, 0.6, 0.25, 0.25])  # Another smaller axes
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax1.plot(x, y, label='Main Plot')
ax2.plot(x, np.cos(x), label='Inset Plot')
ax1.legend()
ax2.legend()
plt.show()
```
